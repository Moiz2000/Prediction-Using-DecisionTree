# Decision Tree

### Task 6-- Prediction Using Decision Tree Algorithm

#### Importing all needed packages for this task 

```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

#### Importing Dataset 

```python
iris_data = pd.read_csv('Iris.csv',index_col=0)

print('Data imported successfully')
iris_data.head(10)
```

#### Getting feature columns fromthe data set

```python
X = iris_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y = iris_data[['Species']]
```

### Setting up the Decision Tree

### We will be using train/test split on our decision tree. Let's import train_test_split

```python
from sklearn.model_selection import train_test_split
```

Now train_test_split will return 4 different parameters. We will name them:
X_trainset, X_testset, y_trainset, y_testset

The train_test_split will need the parameters:
X, y, test_size=0.3, and random_state=3.

The X and y are the arrays required before the split, the test_size represents the ratio of the testing dataset, and the random_state ensures that we obtain the same splits.

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, Y, test_size=0.3, random_state=3)
```

Printing the shape of X_trainset and y_trainset. Ensure that the dimensions match.

```python
print('Shape of X training set {}'.format(X_trainset.shape),'&',' Size of Y training set {}'.format(y_trainset.shape))
```

Printing the shape of X_testset and y_testset. Ensure that the dimensions match.

```python
print('Shape of X training set {}'.format(X_testset.shape),'&',' Size of Y training set {}'.format(y_testset.shape))
```

## Creating Model

```python
Iris_Tree = DecisionTreeClassifier()
```

## Training The Model

We will fit the data with the training feature matrix X_trainset and training response vector y_trainset

```python
Iris_Tree.fit(X_trainset,y_trainset)
```

## Prediction

Let's make some predictions on the testing dataset and store it into a variable called predTree.

```python
predictTree = Iris_Tree.predict(X_testset)
```

We can print out predTree and y_testset to visually compare the predictions to the actual values.

```python
print (predictTree [0:5])
print (y_testset [0:5])
```

## Evaluation

Next, let's import metrics from sklearn and check the accuracy of our model.

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predictTree))
```

## Visualiation

```python
col = iris_data.columns
fig = plt.figure(figsize=(25,20))
plot_tree(Iris_Tree, feature_names=col[[0,1,2,3]], class_names=col[4], filled=True)
```
